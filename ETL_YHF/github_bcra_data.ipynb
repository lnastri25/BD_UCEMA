# Ejemplo consulta APIs con Request

### Ejemplo Github (sin key)

```python
import requests

# URL base de la API de GitHub
base_url = "https://api.github.com"

# Nombre de usuario y nombre del repositorio
username = "MatiasArgentina"
repo_name = "ETL---Clase-7"

# URL completa del repositorio
repo_url = f"{base_url}/repos/{username}/{repo_name}"

# Hacer una solicitud GET a la API de GitHub
response = requests.get(repo_url)

# Imprimir la respuesta JSON
response.json()
```

```python
# Nunca dejamos claves en los repositorios

with open("./Tokens/bcra_token.json", 'r') as archivo:
    contenido = archivo.read()

# Usa la función json.loads() para analizar la cadena de texto JSON en una estructura de datos Python
tokens = json.loads(contenido)
tokens.keys()
```

```python
# LLamo a la API y traigo los resultados
url = 'https://api.estadisticasbcra.com/inflacion_mensual_oficial'

# Token de acceso de la API del BCRA
api_token = tokens['bcra_token']

headers = {"Authorization": f"Bearer {api_token}"}

# Hacer una solicitud GET a la API del BCRA con el encabezado de autorización
response = requests.get(url, headers=headers)
```
